```
#Problema 3.3
"""Este programa calcula la serie 1-1/2+1/3-1/4...+1/n"""
n=0 #número de iteraciones
signo=True#cambiar de signo positivo a negativo  True=positivo
acumulador=0.0
j=1 #contador manual de apoyo
print("Este programa calcula la serie 1-1/2+1/3-1/4...+1/n")
n=int(input("Digita el número de términos de la serie: "))

for i in range(n):
  if signo==True:
    print(f"+(1/{j})")
    acumulador=acumulador+(1/j)
    signo=False
  else:
    print(f"-(1/{j})")
    acumulador=acumulador-(1/j)
    signo=True
  j=j+1

print(f"La suma es: {acumulador}")
 
```

```
#Problema 3.8
"""
Este programa recrea la conjetura de ULAM: 
  La siguiente se llama la conjetura de ULAM en honor del matemático S.Ulam:
    • Comience con cualquier entero positivo
    • Si es par, divídalo entre 2; si es impar, multiplíquelo por 3 y agregúele 1.
    • Obtenga enteros sucesivamente repitiendo el proceso.
  Al final, obtendrá el número 1, independientemente del entero inicial
"""
n=0

n=int(input("Digita cualquier entero positivo: "))

if n>0:
  if n==1:
    print(n)
  while n!=1:
    if (-1)**n > 0:
      n=n/2
      print(n)
    else:
      n=(n*3)+1
      print(n)

else: 
  print("El número tiene que ser positivo ")
```
